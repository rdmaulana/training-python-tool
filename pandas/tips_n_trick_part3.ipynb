# formatting tampilan data frame

```python
n_rows = 5
n_cols = 2
cols = ['operasional', 'omset']

df2 = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), columns=cols)
df2
```

```python
df2['operasional'] = df2['operasional'] * 10000
df2['omset'] = df2['omset'] * 100_000
df2
```

```python
#membuat kolom baru (tanggal)
df2.index = pd.util.testing.makeDateIndex(n_rows, freq='D')
df2 = df2.reset_index()
df2 = df2.rename(columns={'index':'tanggal'})
df2
```

```python
formatku = {
    'tanggal': '{:%d/%m/%y}',
    'operasional': 'Rp {:.2f}',
    'omset': 'Rp {:.2f}'
}

laporan = df2.style.format(formatku)
laporan
```

```python
laporan.hide_index()
laporan.set_caption('Data omset dan operasional')

laporan.highlight_min('omset', color='red')
laporan.highlight_max('omset', color='green')

laporan.highlight_min('operasional', color='grey')
laporan.highlight_max('operasional', color='lightblue')
laporan
```

# merge 2 data frame (kolom)

```python
n_rows = 2
n_cols = 2
cols = ['col1', 'col2']
cols2 = ['col3', 'col4']

df3 = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), columns=cols)
df4 = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), columns=cols2)
```

```python
df3
```

```python
df4
```

```python
df_merge = pd.merge(df3, df4, left_index=True, right_index=True)
df_merge
```

# agregasi dengan groupby

```python
iris_df = pd.read_csv('dataset/iris.csv')
iris_df.head()
```

```python
iris_df.groupby('Species')['PetalLengthCm'].agg(['count', 'mean', 'median'])
```

```python
iris_df.groupby('Species')['PetalLengthCm'].describe()
```

```python
iris_df.groupby('Species')['PetalLengthCm'].count().to_frame()
```

```python
iris_df.groupby('Species')['PetalLengthCm'].mean().to_frame()
```

```python
iris_df.groupby('Species')['PetalLengthCm'].median().to_frame()
```
